```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_csv('USStateData.csv', encoding='windows-1252')
df.head() #dataset was cleaned in Power Query first.
```

```python
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(df.drop(['State'], axis=1))
df_scaled = pd.DataFrame(scaled_features)
```

```python
swingStates = df_scaled.iloc[:5] 
df_scaled = df_scaled[(df_scaled[63] == 0)  | (df_scaled[63] == 1)]
```

```python
df_scaled
```

```python
df_scaled = df_scaled.reset_index()
df_scaled = df_scaled.drop(['index'], axis = 1)
df_scaled
```

```python
X = df_scaled.drop([63], axis = 1)
y = df_scaled[63]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
pca = PCA(n_components = 0.95)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
logreg.score(X_test, y_test) 
```

```python
X_train.shape
```

Model is over fitting. Even if every state was taken into account its unlikely there would be enough observations to make a meaningful impact. 

```python
scores = cross_val_score(logreg, X, y, cv=5)
print(f"Cross-Validation Scores: {scores}")
print(f"Mean Cross-Validation Score: {scores.mean()}")
```
